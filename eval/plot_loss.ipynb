### plot loss curve

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
cm = 1/2.54
```

```python
file_name = ['random_smooth_efno_layer6_4','random_smooth_mlpkan_layer6_4']
# file_name = ['random_smooth_efno_layer6_super_4','random_smooth_mlpkan_layer6_super_4']
# file_name = ['random_smooth_efno_layer6_smalldata_4','random_smooth_mlpkan_layer6_smalldata_4']
n_items = len(file_name)
epoch = []
train_loss = []
test_loss  = []
save_path  = './figs/'
for ii in range(n_items):
    log_file_base  = "../Log/"+file_name[ii]+".log"
    print(log_file_base)
    log_base = np.loadtxt(log_file_base)
    epoch_base   = log_base[:,0]
    train_base   = log_base[:,2]
    test_base    = log_base[:,3]
    epoch.append(epoch_base)
    train_loss.append(train_base)
    test_loss.append(test_base)
```

```python
data = [train_loss,test_loss]
```

```python
plt.rcParams['font.family']       = 'arial' # 'Times New Roman' #
# plt.rcParams['axes.linewidth']    = 1
plt.rcParams['xtick.major.size']  = 2.
plt.rcParams['ytick.major.size']  = 2.5
plt.rcParams['xtick.minor.size']  = 1.5
plt.rcParams['ytick.minor.size']  = 1.5
plt.rcParams['xtick.major.width'] = 0.5
plt.rcParams['ytick.major.width'] = 0.5
plt.rcParams['xtick.minor.width'] = 0.5
plt.rcParams['ytick.minor.width'] = 0.5
plt.rcParams['lines.linewidth']   = 1.5
plt.rcParams['lines.markersize']  = 3.5

plt.rcParams['font.size']    = 10
# plt.rcParams['figure.titlesize'] = 2

plt.rcParams['xtick.labelsize'] = 8
plt.rcParams['ytick.labelsize'] = 8
plt.rcParams['axes.labelsize']  = 10
plt.rcParams['axes.titlesize'] = 8
plt.rcParams['legend.fontsize'] = 8
plt.rcParams['legend.title_fontsize'] = 8

plt.rcParams["savefig.bbox"] = 'tight'
plt.rcParams["savefig.pad_inches"] = 0.1
plt.rcParams['image.cmap'] = 'jet_r'
plt.rcParams['figure.dpi'] = 150
```

```python
title = ['Training', 'Testing']
label= ['EFNO','EFKAN'] #['base','W64','W48','W16','L8','L4','lr=1e-4','lr=5e-3','200']
color      = ['b','#FFA500','c','m','y','k','navy','peru']
linestyle  = 'solid'
linewidth  = 1
marker     = 'None'
markersize = 1
# dpi = 200 
idx0 = 0
idx1 = 200
n_row,n_column = 2,1
figsize=(16*cm,6*cm)
fig,ax     = plt.subplots(n_column,n_row,figsize=figsize)
for ii in range(n_row):
    for jj in range(n_items):
        ax[ii].plot(epoch[jj][idx0:idx1], data[ii][jj][idx0:idx1],color=color[jj], #marker=marker,markersize=markersize,
                linestyle=linestyle, linewidth=linewidth,label=label[jj])
        ax[ii].set_title(title[ii])
    ax[ii].set_yscale("log")
    ax[ii].set_xlabel("epochs")
    ax[ii].set_ylabel(r"Loss")
    ax[ii].legend(loc='upper right')
plt.tight_layout()
plt.savefig(save_path+"loss.pdf",bbox_inches='tight',pad_inches=0.05)
plt.show()
```
